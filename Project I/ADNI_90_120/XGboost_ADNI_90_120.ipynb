```python
from scipy.io import loadmat
import pandas as pd
import numpy as np
from xgboost import XGBClassifier
#from xgboost import predict_proba
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error  
import xgboost as xgb
from sklearn.model_selection import GridSearchCV #网格搜索
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import ensemble
from sklearn.model_selection import train_test_split                 
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
import torch.utils.data as Data
from sklearn import preprocessing
```

```python
m = loadmat("ADNI_90_120_fMRI.mat") 
```

```python
m.keys()
print(m.keys())
print(m['AD'].shape)
print(m['NC'].shape)
print(m['EMCI'].shape)
print(m['LMCI'].shape)
```

```python
#用相关矩阵
AD=np.zeros((m['AD'].shape[0],90,90))
for i in range(m['AD'].shape[0]):
    AD[i]=np.cov(m['AD'][i][:,20:100])
    
NC=np.zeros((m['NC'].shape[0],90,90))
for i in range(m['NC'].shape[0]):
    NC[i]=np.cov(m['NC'][i][:,20:100])
```

```python
plt.imshow(AD[1])
```

```python
#为相关矩阵设置一个阈值
for i in range(m['AD'].shape[0]):
    temp = AD[i].reshape(1,-1)
    AD_index = np.argsort(temp)
    temp[0,AD_index[0,:int(AD_index.shape[1]*0.95)]] = 0
    AD[i] = temp.reshape(90,90)
    
for i in range(m['NC'].shape[0]):
    temp = NC[i].reshape(1,-1)
    NC_index = np.argsort(temp)
    temp[0,NC_index[0,:int(NC_index.shape[1]*0.95)]] = 0
    NC[i] = temp.reshape(90,90)
```

```python
plt.imshow(AD[1])
```

```python
std = preprocessing.MinMaxScaler()

for i in range(AD.shape[0]):
    AD[i] = std.fit_transform(AD[i])
for i in range(NC.shape[0]):
    NC[i] = std.fit_transform(NC[i])

#取出上三角矩阵
cov_AD=np.zeros((m['AD'].shape[0],4005))
#cov_AD=np.zeros((m['AD'].shape[0],90,90))
for i in range(m['AD'].shape[0]):
    temp = []
    for k in range(90):
        temp = temp + AD[i][k][k+1:].tolist()
    #print(np.array(temp).shape)
    cov_AD[i] = temp
    
cov_NC=np.zeros((m['NC'].shape[0],4005))
#cov_AD=np.zeros((m['AD'].shape[0],90,90))
for i in range(m['NC'].shape[0]):
    temp = []
    for k in range(90):
        temp = temp + NC[i][k][k+1:].tolist()
    #print(np.array(temp).shape)
    cov_NC[i] = temp
```

```python
AD = AD.reshape(59,-1)
NC = NC.reshape(48,-1)

AD = pd.DataFrame(AD)
NC = pd.DataFrame(NC)

AD['label'] = [0]*59
NC['label'] = [1]*48

df = AD.append(NC)
label = df['label']
df=df.drop(['label'],axis = 1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(df , label, test_size = 0.1 )  
```

```python
other_params = {'learning_rate': 0.1, 'n_estimators':5, 'max_depth': 2, 'min_child_weight': 3, 'seed': 0,
                    'subsample': 0.9, 'colsample_bytree': 0.6, 'gamma': 0.01, 'reg_alpha': 0.05, 'reg_lambda': 0.5}
 
model=xgb.XGBClassifier(**other_params,eval_metric=['logloss','auc','error'])
train = [x_train, y_train]
eval = [x_test, y_test]
clf = model.fit(x_train, y_train,eval_set=[eval])
pre_train = clf.predict(x_train)
pre_test = clf.predict(x_test)
accuracy_test = accuracy_score(y_test, pre_test)
accuracy_train = accuracy_score(y_train, pre_train)
print("Accuracy_test: %.2f%%" % (accuracy_test * 100.0))
print("Accuracy_train: %.2f%%" % (accuracy_train * 100.0))
```
